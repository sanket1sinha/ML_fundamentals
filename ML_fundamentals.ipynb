```python
import numpy as np
import pandas as pd
headers=["symboling","normalized_losses","make","fuel_type","aspiration","num_doors","body_style","drive_wheels","engine_location","wheel_base","length","width","height","curb_weight","engine_type","num_cylinders","engine_size","fuel_system","bore","stroke","compression_ratio","horsepower","peak_rpm","city_mpg","highway_mpg","price"]
cars=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data",names=headers)
```

```python
cars.head()
```

```python
cars.info()
```

```python
cars=cars.replace('?',np.nan)
```

```python
cars.info()
```

```python
cars[['normalized_losses','horsepower','peak_rpm','price','bore','stroke']] = cars[['normalized_losses','horsepower','peak_rpm','price','bore','stroke']].apply(pd.to_numeric)
```

```python
cars.info()
```

```python
cars[cars['normalized_losses'].isnull()]
```

As we can see that that NaN is present in 41 rows hence we should not delete it rather replace it by average values of that column.

```python
mean_normalized_losses=np.mean(cars["normalized_losses"])
cars['normalized_losses'] = cars['normalized_losses'].replace(np.nan,mean_normalized_losses)
```

Now we have to identifying if more null values are present in our dataframe

```python
cars.isnull().sum()
```

Removing null values from all the other columns

```python
cars=cars.dropna(subset = ['horsepower', 'price', 'num_doors','bore'])
```

```python
cars.info()
```

Now removing all the non numeric columns so that we can normalize our dataset for further calculations

```python
cars=cars.select_dtypes(['number'])
car_price=cars["price"]
```

Normalizing the dataset using maximum and minimum normalization

```python
cars=(cars-cars.min())/(cars.max()-cars.min())
```

```python
cars["price"]=car_price
```

```python
cars.head()
```

Now building a Univariate Model with default value of KNeighborsRegressor

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
%matplotlib inline

def knn_train_test(training_col,target_col,df):
    
    #Splitting the dataframe into training and test dataset
    train_df=df[0:145]
    test_df=df[145:]
    knn=KNeighborsRegressor(n_neighbors=5,algorithm='brute')
    knn.fit(train_df[[training_col]],train_df[target_col])
    predict_price=knn.predict(test_df[[training_col]])
    mse=mean_squared_error(test_df[target_col],predict_price)
    rmse=mse**(1/2)
    return rmse

rmse_list=[]
column_name=list(cars.columns.values)
column_name=column_name[:-1]
for i in column_name:
    rmse_values=knn_train_test(i,"price",cars)
    rmse_list.append(rmse_values)
for i in enumerate(rmse_list):
    print(i)
```

The minimum RMSE is when we use the "curb_weight" column for training

Now building a Univariate Model with multiple values of KNeighborsRegressor

```python
def knn_train_test(training_col,target_col,df,k):
    
    #Splitting the dataframe into training and test dataset
    train_df=df[0:145]
    test_df=df[145:]
    knn=KNeighborsRegressor(n_neighbors=k,algorithm='brute')
    knn.fit(train_df[[training_col]],train_df[target_col])
    predict_price=knn.predict(test_df[[training_col]])
    mse=mean_squared_error(test_df[target_col],predict_price)
    rmse=mse**(1/2)
    return rmse
list_of_rmse=[]
rmse_values=[]
k_list=[1,3,5,7,9]
for i in column_name:
    for j in k_list:
        rmse_per_columm=knn_train_test(i,"price",cars,j)
        rmse_values.append(rmse_per_columm)
    list_of_rmse.append(rmse_values) 
    rmse_values=[]

for i in list_of_rmse:
    x = k_list
    y = list(i)
    plt.plot(x,y)
    plt.xlabel('k value')
    plt.ylabel('RMSE')
    
```

Now building a Multivariate Model with default value of KNeighborsRegressor

```python
def knn_train_test(training_cols,target_col,df):
    
    #Splitting the dataframe into training and test dataset
    train_df=df[0:145]
    test_df=df[145:]
    knn=KNeighborsRegressor(n_neighbors=5,algorithm='brute')
    knn.fit(train_df[training_cols],train_df[target_col])
    predict_price=knn.predict(test_df[training_cols])
    mse=mean_squared_error(test_df[target_col],predict_price)
    rmse=mse**(1/2)
    return rmse

#model with best 2 attributes
cols=["highway_mpg","curb_weight"]
best_2_rmse=knn_train_test(cols,"price",cars)
print("RMSE based on 2 attributes:",best_2_rmse)

#model with best 3 attributes
cols=["highway_mpg","curb_weight","engine_size"]
best_3_rmse=knn_train_test(cols,"price",cars)
print("RMSE based on 3 attributes:", best_3_rmse)

#model with best 4 attributes
cols=["highway_mpg","curb_weight","engine_size","city_mpg"]
best_4_rmse=knn_train_test(cols,"price",cars)
print("RMSE based on 4 attributes:", best_4_rmse)

#model with best 5 attributes
cols=["highway_mpg","curb_weight","engine_size","city_mpg","horsepower"]
best_5_rmse=knn_train_test(cols,"price",cars)
print("RMSE based on 5 attributes:", best_5_rmse)

```

Based on the calcuations made above we should use 3 features for calculating the RSME value.

Now testing our model for different k values

```python
def knn_train_test(training_cols,target_col,df,k):
    
    #Splitting the dataframe into training and test dataset
    train_df=df[0:145]
    test_df=df[145:]
    knn=KNeighborsRegressor(n_neighbors=k,algorithm='brute')
    knn.fit(train_df[training_cols],train_df[target_col])
    predict_price=knn.predict(test_df[training_cols])
    mse=mean_squared_error(test_df[target_col],predict_price)
    rmse=mse**(1/2)
    return rmse
k_list=range(1,26)
list_3_rmse=[]
list_4_rmse=[]
list_5_rmse=[]

for i in k_list:
    cols_3=["highway_mpg","curb_weight","engine_size"]
    rmse_values=knn_train_test(cols_3,"price",cars,i)
    list_3_rmse.append(rmse_values)
plt.plot(k_list,list_3_rmse,label="3 attributes")
plt.xlabel('k value')
plt.ylabel('RMSE')

for i in k_list:
    cols_4=["highway_mpg","curb_weight","engine_size","city_mpg"]
    rmse_values=knn_train_test(cols_4,"price",cars,i)
    list_4_rmse.append(rmse_values)
plt.plot(k_list,list_4_rmse,label="4 attributes")
plt.xlabel('k value')
plt.ylabel('RMSE')

for i in k_list:
    cols_5=["highway_mpg","curb_weight","engine_size","city_mpg","horsepower"]
    rmse_values=knn_train_test(cols_5,"price",cars,i)
    list_5_rmse.append(rmse_values)
plt.plot(k_list,list_5_rmse,label="5 attributes")
plt.xlabel('k value')
plt.ylabel('RMSE')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

From the above plot, best value of RMSE is obtained when k-value is 10 and 4 attributes are used for training the model.
